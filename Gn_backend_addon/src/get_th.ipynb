```python
import pandas  as pd
import json
feature_list = ['18pp','AAC','BPBaac','CTDC','CTDT','CTriad','onehot',
                'PC-PseAAC','ppt25','QSO','SC-PseAAC','CTDD','DPC']#,'ppt'
rate = '1_50'
threshold_dict={}
for model_name in ["XGBClassifier", "GaussianNB", "GradientBoostingClassifier", 
                   "SVC","KNeighborsClassifier", "RandomForestClassifier"]:
    th_dict = {}
    for feature_name in feature_list:
        a = 0
        all_ave = 0
        while a < 5:
            df = pd.read_csv(f'/mnt/md0/Public/一站式/txseml_backend_addon/lib/model/T3/{feature_name}/{rate}/{a}/5fold_results.csv')
            df.set_index(df.columns[0], inplace=True)
            th = df.loc[f'{model_name}','pro_cutoff']
            th = th.replace('[','')
            th = th.replace(']','')
            th = th.split(',')
            
            th = [float(num) for num in th]  # 将字符串列表转换为整数列表
            average = sum(th) / len(th)
            all_ave = all_ave+average
            print(model_name,':',feature_name,':',average)
            a+=1
        th_dict[feature_name] = all_ave/5
    threshold_dict[f'{model_name}']=th_dict
# print(threshold_dict)
with open(f'/mnt/md0/Public/一站式/txseml_backend_addon/lib/model/T3/{rate}_threshold.json', 'w') as json_file:
    json.dump(threshold_dict, json_file)
```
